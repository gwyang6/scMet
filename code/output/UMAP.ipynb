```python
import scanpy as sc
```

```python
adata = sc.read("simulation_adata.h5ad")
adata
```

```python
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)

# PCA
sc.pp.pca(adata)

# UMAP
sc.pp.neighbors(adata)
sc.tl.umap(adata)
sc.pl.umap(adata, color=['celltype'], legend_loc='on data', title='UMAP', frameon=True, show=True)
```
